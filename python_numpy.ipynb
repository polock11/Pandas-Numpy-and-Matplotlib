Advantages of Numpy:
1. Allows Multiple operation 
2. Faster Operations


```python
import numpy as np

#one D array
a = np.array([1,2,3,4])

a.shape

#2D array
two_d = np.array([(1,2,3),(3,4,5)])

# definig data type
c = np.array([(1,2,3),(3,4,5),(4,5,6)], dtype = float)
print(c)
```

```python
# arrays with one particular value

a = np.zeros((3,3))
print(a)

b = np.ones((4,5))
print(b)

c = np.full((5,6), 2)
print(c)
```

```python
# creating identity matriex

a = np.eye(4)
print(a)
```

```python
# create array with random vals

b = np.random.random((3,4)) # all the vals will from 0 to 1
print(b)

c = np.random.randint(10,100,(4,5)) # (start, end, shape)
print(c)

```

```python
# creating array of evenly spaced values --> specifying the value number

a = np.linspace(10,50,5) #(start, end, number_of_vals)
print(a)

# creating array of evenly spaced values --> specifying the step

b = np.arange(10,40,4.5)
print(b)
```

```python
# converting a list to a numpy array
list_one = [1,2,3,4,5,6,7]

np_array = np.asarray(list_one)
print(np_array)
```

Analysing a numpy array

```python
c = np.random.randint(10,90,(5,5))
print(c)

print(c.shape) # returns rows and cloumns
print(c.ndim) # returns dimension
print(c.size) # value count
c.dtype # returns data type
```

#Math Operations

```python
x = np.random.randint(10,30,(3,3))
y = np.random.randint(30,50,(3,3))
print(x+y)
print(x-y)
print(x*y)
print(x/y)
```

#Array Manipulation

```python
c = np.random.randint(10,50,(2,3))
print(c)

trans = np.transpose(c)
print(trans)

d = np.random.randint(10,50,(2,3))
print(d)
trans_one = d.T
print(trans_one)
```
