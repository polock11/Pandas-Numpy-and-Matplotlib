```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
# ploting the data

# sin wave
plt.plot(x,y)
plt.xlabel('angel')
plt.ylabel('sin_value')
plt.title('sin wave')
plt.show()
```

```python
# cos wave
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cos value')
plt.title('Cos Wave')
plt.show()
```

```python
# ploting a parabola 
x = np.linspace(-10,10,20)
y = x**2
z = x**3
plt.plot(x,y,'r--') # 'r.' --> r for red, '.' for plot symbol
plt.plot(x,z, 'g-')
plt.show() 
```

```python
# ploting multile line in one graph
a = np.linspace(-5,5,10)
plt.plot(a,np.sin(a),'r-')
plt.plot(a,np.cos(a),'g--')
plt.show()
```

```python
#Bar plots
fig = plt.figure()
axis = fig.add_axes([0,0,1,1]) 
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
parcentage = [100,30,40,60,10]
axis.bar(languages, parcentage)
plt.xlabel('Language')
plt.ylabel('Percentage')
plt.show()
```

```python


languages = ['English', 'French', 'Spanish', 'Latin', 'German']
parcentage = [100,30,40,60,10]
plt.pie(parcentage, labels = languages, autopct = '%1.1f%%')
```

```python
#Pie chart
fig2 = plt.figure()
axis = fig2.add_axes([0,0,1,1])
axis.pie(parcentage, labels = languages, autopct = '%1.1f%%')
plt.show()
```

```python
#scatter plot
x = np.linspace(0,10, 30)
y = np.sin(x)
z = np.cos(x)
fig3 = plt.figure()
axis = fig3.add_axes([0,0,1,1])
axis.scatter(x,y,color='g')
axis.scatter(x,z,color='b')
plt.show()
```
