#Array Operations

```python
x = np.array([1,2,3,4,5,6])
y = x**2
plt.plot(x,y)
plt.show()
```

```python
def f(x):
    return x**2/np.exp(x)+np.sin(x) #elementwise operation

x = np.linspace(-5,5,100)
y = f(x)
plt.plot(x,y)
```

```python
theta = np.linspace(0,2*np.pi,1000)
r = 1 + 3/4 * np.sin(3*theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.plot(x,y)
```

```python
A = 1/2 * sum((r**2) * (theta[2]-theta[1]))
print(A)
```

```python
arc_length = sum(np.sqrt((r**2)+(np.gradient(r,theta)**2)) * theta[1]-theta[0])
print(arc_length)
```

# Solve System of equations

3x+2y+z=4,
5x-5y+4z=3,
6x+z=0

```python
# Solve System of equations
co_eff = np.array([[3,2,1],[5,-5,4],[6,0,1]])
val = np.array([4,3,0])
np.linalg.solve(co_eff,val)
```
